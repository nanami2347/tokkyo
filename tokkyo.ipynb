```python
import os

dir_path = '/content/drive/My Drive/tokkyo'

files = os.listdir(dir_path)
print(files)

print(type(files))
```

```python
import csv
import pprint
```

```python
from __future__ import print_function
import re

list1 = []
gakusyu_list1 = []
gakusyu_list2 = []

for a in files:
  path = '/content/drive/My Drive/tokkyo' + '/' + a
  f = open(path)
  f.close()

  with open(path) as f:
    s = f.read()

  sentence = s
  waketa = sentence.split('【')

  #gakusyu_list1 = []
  #gakusyu_list2 = []


  list = []
  for a in waketa:
    b =  '【' + a
    list.append(b)

    pattern_a = '考案が解決しようとする課題'
    pattern_b = '課題を解決するための手段'

  for tango in list:
    #gakusyu_list=[]
    search_result = re.search(pattern_a,tango)
    #print("result:",search_result)
    if search_result:
      kadai = tango.replace('【考案が解決しようとする課題】', '')
      kadai_1 = kadai.replace('<br/>\n\n', '')
      #print("kadai:",kadai_1)
      gakusyu_list1.append(kadai_1)
    search_result = re.search(pattern_b,tango)
    if search_result:
      syudan = tango.replace('【課題を解決するための手段】', '')
      syudan_1 = syudan.replace('<br/>\n\n', '')
      gakusyu_list2.append(syudan_1)
#print(gakusyu_list1,gakusyu_list2)
gakusyu_list=[]
for kadai,syudan in zip(gakusyu_list1,gakusyu_list2):
  gakusyu_list.append([kadai,syudan])

print(gakusyu_list)
```

```python
with open('/content/drive/My Drive/tokkyo.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(["instruction","output"])
    writer.writerows(gakusyu_list)
```

```python
import json

with open("/content/drive/My Drive/tokkyo.csv", 'r') as f:
    csv_data = csv.DictReader(f)
    json_data = json.dumps([row for row in csv_data], ensure_ascii=False)

with open('/content/drive/My Drive/tokkyo.json', 'w') as json_file:
    json_file.write(json_data)
```
